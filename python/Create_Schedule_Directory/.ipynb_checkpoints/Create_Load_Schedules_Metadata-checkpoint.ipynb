```python
import snowflake.connector as snow
import pandas as pd
```

```python
username= "your-username"
password = "your-password
account= "your-account"
role = "your-role"
warehouse = "your-warehouse
database = "your-database"
schema = "your-schema
```

```python
def main(username, password, account, role, warehouse, database, schema):
    try: 
        conn = snow.connect(
        user= username,
        password = password,
        account = account,
        role = role,
        warehouse = warehouse,
        database = database,
        schema = schema)
        cur = conn.cursor()
        sql = f"ALTER WAREHOUSE {warehouse} RESUME IF SUSPENDED;"
        cur.execute(sql)
        print("Spinning up Warehouse")
        sql = '''Create table if not exists Schedule_Metadata (
        Mnemonic VARCHAR, 
        Item_Code VARCHAR, 
        MDRM_Code VARCHAR,
        Start_Date DATE, 
        End_Date DATE, 
        Item_Name VARCHAR,
        Confidentiality VARCHAR, 
        Item_Type VARCHAR, 
        Form_Type VARCHAR,
        Description VARCHAR, 
        Series_Glossary VARCHAR, 
        Schedule_Type VARCHAR, 
        Schedule_Label VARCHAR, 
        MDRM_Label VARCHAR, 
        Data_Type VARCHAR);'''
        cur.execute(sql)
        sql = 'Create Temporary Stage schedule_metadata_stage;'
        cur.execute(sql)
        print('Stage Created')
        sql = r'Put file:///{insert_file\path\here}\call-report-database\MDRM_CSV_Filtered_Schedules_Data_Types.csv @schedule_metadata_stage'
        cur.execute(sql)
        print('File put in stage')
        sql = r"Copy Into Schedule_Metadata From @schedule_metadata_stage on_error=skip_file FILE_FORMAT=(SKIP_HEADER=1 ERROR_ON_COLUMN_COUNT_MISMATCH=FALSE EMPTY_FIELD_AS_NULL =TRUE FIELD_OPTIONALLY_ENCLOSED_BY='\"' TRIM_SPACE=TRUE);"
        cur.execute(sql)
        print('Data copied to table')#need to figure out way to copy error messages to here
        conn.commit()
        sql = "Select Mnemonic, Form_Type, Schedule_Type From Schedule_Metadata where Mnemonic = 'RCFA';"
        returns = cur.execute(sql)
        for rets in returns:
            print(rets)
        sql = f"ALTER WAREHOUSE {warehouse} SUSPEND;"
        cur.execute(sql)
        print('Warehouse Suspended')
    except Exception as e: 
        print(e)
    finally:
        if conn: 
            conn.close()
            print('Snowflake Connection Closed')
```

```python
if __name__ == "__main__":
    main(username,password,account, role, warehouse,database, schema)
```
